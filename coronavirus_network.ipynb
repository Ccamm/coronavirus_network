```python
# Add compatibility with Windows
DATASET_URLS_CSV = "dataset/dataset_urls.csv"

def load_datasets_to_dataframe(dataset_urls=DATASET_URLS_CSV):
    column_names = ("dataset_label", "url")
    df_urls = pd.read_csv(dataset_urls, header=None, names=column_names)
    datasets = {}
    
    for index, row in df_urls.iterrows():
        data_label = row[column_names[0]]
        url = row[column_names[1]]
        dwnld = requests.get(url).content
        data_df = pd.read_csv(io.StringIO(dwnld.decode('utf-8').replace('\r', '')))
        datasets[data_label] = data_df
        
    return datasets
        
    
datasets = load_datasets_to_dataframe()
```

```python
latest_date = datasets['confirmed'].columns[-1] # number of confirmed cases to date
df = datasets['confirmed'].head(265)[['Country/Region', 'Province/State', 'Lat', 'Long', latest_date]]
df 
```

```python
# This function generates a graph given the dataframe
pos = {}
# df - dataframe containing the data, pymap - map to be given to, G- graph, scale - scale of the number of cases
def generateGraph(df, pymap, G, scale=1):
    for index, row in df.iterrows():
        coords = (row['Lat'], row['Long'])
        name = row['Country/Region']

        # Adds the state or province if the point has one
        if not pd.isnull(row['Province/State']):
            name += ' - ' + row['Province/State']
            
        # Add number of cases in the title
        name += ' - Confirmed:' + str(row[latest_date])
            
        # Add to pos dict for drawing
        pos[name] = coords

        # Adds the nodes to the graph
        G.add_node(name, pos=coords) 
        # Marker
        marker = Marker(location=coords, draggable=False, title=name, opacity=0.1)
        pymap.add_layer(marker)
        # Circle
        circle = Circle(location=coords, radius=round(row[latest_date]*scale), color='red', fill_color='red')
        pymap.add_layer(circle)
```

```python
# Generate map for the confirmed cases
confirmedCasesGraph = nx.Graph()

confirmedMap = Map(basemap=basemaps.OpenStreetMap.Mapnik, center=center, zoom=zoom)
confirmedMap.add_control(FullScreenControl())
confirmedMap.add_control(LayersControl())

generateGraph(df, confirmedMap, confirmedCasesGraph, 1.2)
```
