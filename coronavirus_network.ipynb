```python
import networkx as nx
import matplotlib as plt
import numpy as np
import pandas as pd
import math, random, requests, io
%matplotlib inline
```

```python
# Add compatibility with Windows
DATASET_URLS_CSV = "dataset/dataset_urls.csv"

def load_datasets_to_dataframe(dataset_urls=DATASET_URLS_CSV):
    column_names = ("dataset_label", "url")
    df_urls = pd.read_csv(dataset_urls, header=None, names=column_names)
    datasets = {}
    
    for index, row in df_urls.iterrows():
        data_label = row[column_names[0]]
        url = row[column_names[1]]
        dwnld = requests.get(url).content
        data_df = pd.read_csv(io.StringIO(dwnld.decode('utf-8').replace('\r', '')))
        datasets[data_label] = data_df
        
    return datasets
        
    
datasets = load_datasets_to_dataframe()
```
